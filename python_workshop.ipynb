### circle coloring

```python
y=np.array([22,33,66,77,66])
plt.pie(y)
plt.show()
```

```python
y=np.array([22,33,66,77,66,444,77])
plt.pie(y,labels=["RTC","DTC","NTC","VTC","STC","PTC","GTC"],autopct="%2.1F%%")
plt.show()
```

```python
y=np.array([22,33,66,77,66,444,77])
plt.pie(y,labels=["RTC","DTC","NTC","VTC","STC","PTC","GTC"],autopct="%2.1F%%")
plt.legend(loc="best")
plt.legend(loc="upper right")
plt.show()
```

```python
labels='A','B','C','D','E','F'
sizes=[15,20,10,17,1,37]
explode=(0,0.1,0,0,0,0)
plt.figure()
plt.pie(sizes,labels=labels,explode=explode,autopct='%1.1F%%')
plt.axis('equal')
plt.show()
```
